```python
#This technique would combine all reads before performing Trinity

'''IMPORT SUBPROCESS'''
import subprocess

'''COMBINE READS'''
#NOTE: For a single-read run, one Read 1 (R1) FASTQ file is created for each sample per flow cell lane. 
      #For a paired-end run, one R1 and one Read 2 (R2) FASTQ file is created for each sample for each lane.
#If single-read run:


#If paired-end run:
#Write functions to concatenate forward and reverse reads in paired en case:
def combine_reads(reads_pattern, output_file):
    """
    Combine reads matching a pattern into a single file. So all R1 in one file and all R2 in the other
    Parameters:
    reads_pattern (str): Glob pattern for matching read files.
    output_file (str): Name of the output file.
    """
    # Using shell command to concatenate files
    cat_command = f"cat {reads_pattern} > {output_file}"
    subprocess.run(cat_command, shell=True, check=True)
    print(f"Combined reads into {output_file}")

 

'''COMBINE EVERYTHING IN MAIN FUNCTION'''

def main():
    # Define your read patterns and output files
    forward_pattern = "*_1.fastq" #Here basically all the R1 will technically be denoted _1.fastq while R2 will be denoted _2fastq
    reverse_pattern = "*_2.fastq"
    combined_forward = "combined_1.fastq"
    combined_reverse = "combined_2.fastq"
    
    # Combine reads
    combine_reads(forward_pattern, combined_forward)
    combine_reads(reverse_pattern, combined_reverse)
    
    # Optionally, run FastQC on the combined files
    # Uncomment the next line to enable FastQC check
    # run_fastqc([combined_forward, combined_reverse])
    
    # Run Trinity
    run_trinity(combined_forward, combined_reverse)

if __name__ == "__main__":
    main()
    
'''RUN TRINITY'''

def run_trinity(forward_reads, reverse_reads, memory="50G", cpu="10"):
    """
    Run Trinity for de novo transcriptome assembly.
    
    Parameters:
    forward_reads (str): File name of combined forward reads.
    reverse_reads (str): File name of combined reverse reads.
    memory (str): Maximum memory for Trinity to use.
    cpu (str): Number of CPU threads for Trinity.
    """
    trinity_command = f"Trinity --seqType fq --max_memory {memory} --left {forward_reads} --right {reverse_reads} --CPU {cpu}"
    subprocess.run(trinity_command, shell=True, check=True)
    print("Trinity assembly completed")

'''EXECUTE'''
```

```python
import os
import glob

def concatenate_fasta_files(output_file, *input_directories):    
    """
    Concatenate FASTA files from different conditions into a single file for proper de novo assembly with Trinity

    Parameters:
    output_file (str): Path to the output FASTA file.
    *input_directories (str): Variable number of arguments, each specifying a directory containing FASTA files.
    Example: concatenate_fasta_files(output_fasta,
                        'path/to/condition1_fasta_files',
                        'path/to/condition2_fasta_files',
                        'path/to/condition3_fasta_files',
                        'path/to/condition4_fasta_files')
    """
    # Open the output file
    with open(output_file, 'w') as outfile:
        # Process each directory of FASTA files
        for directory in input_directories:
            # Create a list of FASTA files in the current directory
            fasta_files = glob.glob(os.path.join(directory, '*.fasta'))
            print(f"Processing {len(fasta_files)} files in {directory}") #This line will confirm to the user the amount of files that is being processed
            
            # Iterate through each FASTA file
            for file in fasta_files:
                with open(file, 'r') as infile:
                    # Write each line from the input file to the output file so to concatenate all the FASTA files in all the different conditions. 
                    outfile.write(infile.read())
```
